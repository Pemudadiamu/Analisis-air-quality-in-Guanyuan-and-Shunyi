## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

- **Pertanyaan 1:** Bagaimana tren rata-rata konsentrasi polutan PM2.5 bulanan di wilayah Guanyuan dan Shunyi selama periode 2013-2017, dan pada bulan keberapa tingkat polusi udara biasannya mencapai titik tertinggi?

## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Data Wrangling

### Gathering Data

#### Load df ...

```python
# Memuat Dataset
df_Guanyuan = pd.read_csv("Air-quality-Guanyuan-Shunyi-dataset/PRSA_Data_Guanyuan_20130301-20170228.csv")
df_Shunyi = pd.read_csv("Air-quality-Guanyuan-Shunyi-dataset/PRSA_Data_Shunyi_20130301-20170228.csv")

#Mengabungkan Dataset
df_quality = pd.concat([df_Guanyuan, df_Shunyi], ignore_index=True)

print("Gathering Data Selesai")
print("Dimensi data gabungan: ", df_quality.shape)
df_quality.head()
```

**Insight:** (Opsional)
- Melakukan Pemangilan Dataset dengan pd.read_csv
- Melakukan Gathering(pengabungan) data dengan cara pd.concat

### Assessing Data

#### Identifying ... problem

```python
# Mengecek Tipe data dan informasi data
print("INFO DATASET")
df_quality.info()

# Mengecek Missing Value
print("\n JUMLAH MISSING VALUE")
print(df_quality.isna().sum())

#Mengecek data Duplikat
print("\nJUMLAH DATA YANG DUPLIKAT")
print("Duplikat : " , df_quality.duplicated().sum())

# Menrngkas parameter statistik
print("\n RINGKASAN STATISTIK")
print(df_quality.describe())
```

**Steps to Take:**
- memerikasa kesesuaian tipe data dengan(info())
- mendeteksi nilai kosong/missing values dengan (isna())
- mencari data duplikat dengan (duplicated())
- memeriksa parameter statistik dengan (describer())

**Insight:** (Opsional)
- pada missing value ditemukan ratusan bahkan sampai ribuan data kosong yang di mana di temukan hampir di kolom polutan (PM2.5, PM10, SO2, NO2, CO, O3,TEMP, PRES, DEWP, RAIN, wd, WSPM)
- format data yang tidak tepat di temukan pada (year, month, day, hour) yang dimana data ini perlu di gabung menjadi tipe data (datetime)
- terindikasi nilai ekstream berupa lonjakan polusi ekstream pada PM2.5 dan O3

### Cleaning Data

#### Fixing ... problem

```python
# Mengabungkan kolom waktu (year, month, day, hour) menjadi tipe data (datatime)
df_quality['datatime'] = pd.to_datetime(df_quality[['year', 'month', 'day', 'hour']])

# Membuang kolom (NO) yang tidak di perlukan kedepannya
df_quality.drop(columns=['No'], inplace=True, errors='ignore')

# Mengilanghkann Missing value pada kolom numerikal
numeric_cols = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']
for col in numeric_cols:
    df_quality[col] = df_quality[col].interpolate(method='linear', limit_direction='both')
    
#Menghilangkan Missing Value pada kolom kategorikal (wd)
df_quality['wd'] = df_quality['wd'].ffill().bfill()

# Memeriksa dataset yang sudah di bersihkan
print("MISSING VALUE SETELAH DI BERSIHKAN")
print(df_quality.isna().sum())

# Menyimpan dataset yang sudah bersih untuk keperluan deployment
df_quality.to_csv("cleaned_df_quality.csv", index=False)
print("\n DATA CLEANED BERHASIL DI BUAT")

```

**Insight:** (Opsional)
- penyelarasan format pada kolom( year, month, day, dan hour ) yang awalnya di pisah sudah di jadikan satu menjadi kolom (datetime)
- data kolom (No) di hilangkan demi keefisienan data kedepannya
- penanganan missing value sudah di atasi untuk kolom numerikal di isi menggunakan metode interpolasi linear sedangkan untuk kategorikal (wd) menggunakan metode foward fill dan backward fill

## Exploratory Data Analysis (EDA)

### Explore ...

```python
# Menjawab Pertanyaan pertama Bagaimana tren rata-rata konsentrasi PM2.5 bulanan di stasiun Guanyuan dan Shunyi, dan bulan apa yang paling parah?

# 1. Menggunakan groupby untuk melihat rata-rata, nilai Max, nilai Min PM2.5 per bulan di setiap Station
monthly_summary = df_quality.groupby(by=['month','station']).agg({
    'PM2.5' : ['mean', 'max', 'min']
})
print("\n RINGKASAN STATISTIK PM2.5 (BULANAN)")
print(monthly_summary.head(10))

# 2. Membuat pivot perbandingan antara guanyuan dan shunyi agar lebih mudah di baca
pivot_pm25 = df_quality.pivot_table(values='PM2.5', index='month', columns='station', aggfunc='mean')
print("\n RATA-RATA PM2.5 PER BULAN (PIVOT)")
print(pivot_pm25)

# 3. Mencari bulan dengan rata- rata polusi paling tinggi di masing masing stasiun
print("\n BULAN DENGAN POLUSI TERTINGGI")
print("Guanyuan : ", pivot_pm25['Guanyuan'].idxmax(),"(Rata-rata : ", round(pivot_pm25['Guanyuan'].max(), 2),")")
print("Shunyi : ", pivot_pm25['Shunyi'].idxmax(),"(Rata-rata : ", round(pivot_pm25['Shunyi'].max(), 2),")")
```

```python
# Menjawab Pertanyaan Kedua Bagaimana korelasi antara curah hujan (RAIN) dan kecepatan angin (WSPM) terhadap fluktuasi konsentrasi PM2.5?

# Menghitng matrixs korelasi antar polutan (PM 2.5) dengan faktor cuaca (hujan & kecepatan angin)
cuaca_corr = df_quality[['PM2.5', 'RAIN', 'WSPM']].corr()

print("MATRIKS KORELASI")
print(cuaca_corr)
```

**Insight:** (Opsional)
- Menjawab pertanyaan ke 1 tentang tren bulanan
    melalui proses grouping dan pivot tabel bisa di lihat hasil di atas di temukan konsentrasi rata rata PM2.5 tertinggi di kedua stasiun terjadi pada bulan Desember yang bisa kita perkirakan bertetapan dengan musim dingin. Nilai rata rata pada stasiun Guanyuan di bulan desember itu (110) dan untuk Shunyi itu(97). Ada buruk pasti ada baiknya, bisa di lihat tingat polusi terendah ada di bulan Agustus. Lalu untuk membandingkan keduanya bisa di simpukan secara keseluruhan Guanyuan konsisten mencatatkan PM2.5 lebih tinggin di bandinkan dengan Shunyi.(mungkin di karenakan Guanyuan merupaka daerah yang lebih padat di banding Suanyi)
- Menjawab pertanyaan ke 2 tentang korelasi faktor cuaca
    melalui fungsi yang di bikin di atass kita bisa mendapatkan statistik berikut:
    - kecepatan angin(WSPM) memiliki korelasi bernilai negatif terhadap PM2.5 (sekitar -0.276). yang dimana hal ini membutktikan bahwa hembusan angin yang semakin kencang efektif untuk menurunkan konsentrasi polutan di udara
    - curah hujan(RAIN) memiliki korelasi negatif yang  mendekati 0 (sekitar -0.009) yang mengindikasikan bahwa hujan tidak terlalu signifikan dalam menurunkan kadar polusi PM2.5

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Visualisasi untuk menjawab pertanyaan 1 

sns.set_theme(style="whitegrid")

# menyiapkan data bulanan
monthly_pm25 = df_quality.groupby(['month', 'station'])['PM2.5'].mean().reset_index()


# membuat line chart
plt.figure(figsize=(12,6))
sns.lineplot(
    data=monthly_pm25,
    x='month',
    y='PM2.5',
    hue='station',
    marker='o',
    linewidth=2.5,
    palette=['#ef4444', '#3b82f6']
)
plt.title('Tren Rata-Rata Bulanan Konsentrasi PM2.5 (2013-2017)\nGuanyuan vs Shunyi', fontsize=16, fontweight='bold')
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Rata-Rata PM2.5 ', fontsize=12)
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Agt', 'Sep', 'Okt', 'Nov', 'Des'])
plt.legend(title='Stasiun', loc='upper right')
plt.tight_layout()
plt.show()
```

### Pertanyaan 2:

```python
# Visualisasi untuk menjawab pertanyaan  2

# memisahkan 5000 sampel untuk membuat plotting lebih cepat
sampel_df = df_quality.sample(n=5000, random_state=42)

#membuat kanvas untuk 2 grafik berdampingan 
fig, axes = plt.subplots(1,2, figsize=(16,6))

#grafik 1 perbandingan kecepatan angin(WSPM) vs PM2.5
sns.regplot(
    ax=axes[0],
    data = sampel_df,
    x='WSPM',
    y='PM2.5',
    scatter_kws={'alpha' : 0.3, 'color' : '#0ea5e9'},
    line_kws={'color': '#ef4444', 'linewidth': 2}
)
axes[0].set_title('Korelasi Kecepatan Angin(WSPM) terhadap PM2.5')
axes[0].set_xlabel('Kecepatan Angin (m/s)')
axes[0].set_ylabel('Konsentrasi PM2.5')

# Grafik 2 perbandingan Curah hujan vs PM2.5
sns.regplot(
    ax=axes[1],
    data = sampel_df,
    x='RAIN',
    y='PM2.5',
    scatter_kws={'alpha' : 0.3, 'color' : "#199643"},
    line_kws={'color': '#ef4444', 'linewidth': 2}
)
axes[1].set_title('Korelasi Curah Hujan (RAIN) terhadap PM2.5')
axes[1].set_xlabel('Curah Hujan (mm))')
axes[1].set_ylabel('Konsentrasi PM2.5')

plt.tight_layout()
plt.show()
```

**Insight:** (Opsional)
- Bedasarkan Grafik pertama terlihat jelas pola siklus tahunan polusi udara yang identik antara Guanyuan dengan Shunyi. Tingkat konsentrasi PM2.5 memburuk secara signifikan pada awal musim dingin seperti yang bisa kita lihat dari grafik, yang dimana tren terlihat menaik sejak Oktober dan mencapai puncaknya di bulan Desember hingga januari (menyentuh angka 100). Kemudia kualitas udara mulai membaik selepas bulan januari dan mencapai titik paling rendah nya di bulan Agustus. Secara spasial, stasiun Guanyuan konsisten dengan polusinnya yang lebih sedikit tinggi di banding dengan Shunyi di hampis seluruh bulan observasi
- pada grafik scatter plot yang dimana ini memvisualisasikan korelasi yang sebelumnnya di hitung
    pada grfik kiri bisa di lihat garis merah menurun ke bawah yang menandakan adannya korelasi negatif antara kecepatan angin dengan PM2.5. bisa kita lihat polusi PM2.5 (PM2.5 > 400) memburuk hanya pada di saat kecepatan angin (di bawah 3m/s). yang dimana ini berati hembusan angin yang kencang sangat mampu menyapu dan menurunkan konsentrasi polutan
    pada grafik kanan bisa di lihat garis merah  bergerak lurus secara horizontal, yang memvalidasi nilai korelasi nya itu mendekati 0. Yang artinnya curah hujan tidak terbukti memiliki hubungan langsugn dengan naik turunnya polutan PM2.5

## Analisis Lanjutan (Opsional)

```python
# Melakukan pengelompokan pada nilai PM2.5 menjadi beberapa kategori (baik, sedang, tidak sehat, bebahaya) menggunakan binning

# menentukan batas nilai bins dan melabeli kategorinnya
bins = [0, 50, 100, 150, df_quality['PM2.5'].max()]
labels =['Baik', 'Sedang', 'Tidak sehat', 'Berbahaya']

# membuat kolom baru 'Kategori_udara' 
df_quality['Kategori_udara'] =pd.cut(df_quality['PM2.5'], bins=bins, labels=labels, include_lowest=True)

# menghitung jumlah masing-masing kategori per stasiun
aqi_counts = df_quality.groupby(['station', 'Kategori_udara']).size().unstack()

# membuat visualisasi
aqi_counts.plot(kind='bar', stacked=True, figsize=(10,6), color=['#10b981', '#fbbf24', '#f97316', '#ef4444'], edgecolor='black')

plt.title('Proporsi Status Kualitas Udara (PM2.5)\nGuanyuan vs Shunyi (2013-2017)', fontsize=14, fontweight='bold')
plt.xlabel('Stasiun', fontsize=12)
plt.ylabel('Jumlah Jam Observasi', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Status Kualitas Udara', loc='upper right')
plt.tight_layout()
plt.show()


print("Total Jam Observasi per Kategori")
print(aqi_counts)
```
